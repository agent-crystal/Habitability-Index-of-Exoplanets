```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_excel('exoplanetss.xlsx')
df['lum_watt'] = df['st_lum']*(3.826*10**26) # converting solar luminosities to watts
df['semi_major_meters'] = df['pl_orbsmax']*(1.496*10**11) # converting au to meters
df
```

```python
pi = 3.14159265359
df['is_terrestrial'] = np.where(df['st_logg'] < 9.8, 1, 0)# compares planet's surface gravity to that of Earth
flux = df['lum_watt']*(1-0.2)  #0.2 is taken as albedo here due to its value being slightly lower than Earth's (0.3) 
denom = 16*pi*df['semi_major_meters']**2*((1 - df['pl_orbeccen']**2)**0.5)
df['flux']= flux/denom
```

```python
rows = df[(df['flux']>=65) & (df['flux']<=300)].copy()
rows['habitability_index'] = rows['flux']*rows['is_terrestrial']
rows
```

```python
plt.scatter(rows['habitability_index'], rows['sy_dist'],marker='.', c=rows['habitability_index'], cmap='magma')
plt.colorbar(label='Habitability Index')
plt.xlabel('Habitability index')
plt.ylabel('Distance/pc')
plt.title('Habitability of Exoplanets')
plt.show()
```
